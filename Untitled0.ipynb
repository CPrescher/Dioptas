```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plot=plt.plot

index = np.arange(1,1000)
d=79e-6
r=200e-3

twotheta=np.arctan(index*d/r)
twotheta_step=twotheta
twotheta_step[1:]=twotheta[1:]-twotheta[:-1]


plt.plot(index,twotheta_step/np.max(twotheta_step))
```

```python
x=index*d
y=twotheta_step/max(twotheta_step)
plot(x,y)
```

```python
from scipy.optimize import curve_fit

def func(x, a):
    return np.exp(-x**2/(2*a**2))

popt, pcov = curve_fit(func, x,y)

print popt
```

```python
plot(x,y)
plot(x, func(x,popt))
```
